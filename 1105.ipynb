```python
import requests,codecs,csv,io,json
from bs4 import BeautifulSoup

r1=requests.get('https://www.cdc.gov.tw/Category/NewsPage/EmXemht4IT-IRAPrAnyG9A')

b1=BeautifulSoup(r1.text,  "html.parser")
b2=b1.find_all("tr")
fn=0
for d1 in b2:
  if d1.find("td")!=None:
    r2=requests.get("https://www.cdc.gov.tw"+d1.find_all("td")[0].find("a").attrs["href"])
    b3=BeautifulSoup(r2.text, "html.parser")
    fn+=1
    with codecs.open("news/"+str(fn)+".txt","w",r2.encoding) as f:
      f.write(d1.find_all("td")[1].text+"\r\n")
      f.write(d1.find_all("td")[0].text.strip(" \r\n\t")+"\r\n")
      f.write(b3.find("p",{"class":"con-word"}).text.strip(" \r\n\t"))
    # print(d1.find_all("td")[1].text,d1.find_all("td")[0].text.strip(" \r\n\t"))
  # print(d1.text)
  # print(d1.parent.find("p",{"class":"time"}).text)
  # print(d1.parent.find("p").text)
  # print("===========================")
```
