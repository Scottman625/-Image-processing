```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from PIL import ImageFont, ImageDraw, Image
import time
def f1(x):
  m1=np.full((300, 500, 3),(200,255,200), np.uint8)
  cv2.rectangle(m1, (x,100), (x+90,200), (255,0,0), -1)
  dp.clear_output(True)
  p1.write(m1)
p1=cv2.VideoWriter("2.mp4", cv2.VideoWriter_fourcc(*'MP4V'), 60,(500,300))
count=10
for n in range(5000):
  for i in range(10,400,1):
    count+=1
    if count <=398:
      f1(i)
    else:
      for i in range(400,10,-1):
        count -= 1
        if count >=12:
          f1(i)
        else:
          continue
  
p1.release()
```
